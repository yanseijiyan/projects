THIS APP CALCULATE THE REFUND OF BOTTLES CONTAINERS

```python
Less_deposit = 0.12
More_deposit = 0.27

less = int(input('How many containers 1 litter or less? :'))
more = int(input('How many containers more thean 1 litter? :'))

refund = less * Less_deposit + more * More_deposit
print('Your total refund will be $%.2f'%refund)
```
